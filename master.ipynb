### Model Training

```python
# Model Base
def train_model (model, x_train, y_train):
  model.fit(x_train, y_train)
  return model

def model_predict (model, X_test):
  y_pred = model.predict(X_test)
  return y_pred

# Evalution Performance
def eval (model,x_train,y_train,y_test, y_pred):
    print(f"Training Accuracy: {model.score(x_train, y_train)}")
    print(f"Testing Accuracy: {accuracy_score(y_test, y_pred)}")
    cm = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix:")
    print(np.array2string(cm, separator=', '))
    print("\n" + "="*40 + "\n")

models = {
    'SVC': SVC(kernel='linear'),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
    'MultinomialNB': MultinomialNB()
}    
  
```

```python
for model_id, modeling in models.items():
    print (f"Results for {model_id}")
    train_model(modeling, x_train, y_train)
    pred = model_predict(modeling, x_test)
    eval(modeling, x_train, y_train, y_test, pred)
```

## Choosing SVM & Checker

## Rec SYS

```python
def customizing (dis):
    desc = description[description['Disease'] == dis]['Description']
    desc = " ".join([w for w in desc])

    pre = precautions[precautions['Disease'] == dis][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
    pre = [col for col in pre.values]

    med = medications[medications['Disease'] == dis]['Medication']
    med = [med for med in med.values]

    die = diets[diets['Disease'] == dis]['Diet']
    die = [die for die in die.values]

    wrkout = workout[workout['disease'] == dis] ['workout']


    return desc,pre,med,die,wrkout
```

```python
def rec (patient_symptoms):
    vector = np.zeros(len(symptoms_idecies))
    for item in patient_symptoms:
        vector[symptoms_idecies[item]] = 1
    return diseases_idecies[svc.predict([vector])[0]]
```
